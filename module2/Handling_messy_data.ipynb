```python
data = {
    "Brand": ["Maruti", "Hyundai", "Maruti", "Honda", "Toyota", "Honda"],
    "Year": [2018, 2017, 2018, 2016, 2019, 2015],
    "Fuel": ["Petrol", "Diesel", "Petrol", "Petrol", "Diesel", "Petrol"],
    "KMs_Driven": [30000, 40000, 30000, None, 20000, 70000],
    "Price": [400000, 500000, 400000, 450000, 800000, -200000],
}
```

TASK-1

```python
import numpy as np
import pandas as pd
```

```python
df= pd.DataFrame(data)
```

```python
df.head()
```

```python
df.isnull().sum()
```

TASK-2

Filling missing values

```python
df['KMs_Driven']= df['KMs_Driven'].fillna(df['KMs_Driven'].median())
```

```python
df= df[df['Price']>0]
```

```python
df= df.drop_duplicates()
```

```python
df
```

TASK-3

```python
mapping= {'Petrol':1,'Diesel':2}
df['Fuel']= df['Fuel'].map(mapping)
```

```python
df= pd.get_dummies(df, columns=['Brand']).astype(int)
```
