```python
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=10, random_state=42)
```

```python
print(x[10])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

TASK-1 data leakage

```python
scale= StandardScaler()
```

```python
x_scaled= scale.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
lr= LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_train= lr.predict(x_train)
```

```python
y_pred_test= lr.predict(x_test)
```

```python
print("Training_Accuracy_score: ", accuracy_score(y_pred_train,y_train))
print("Test_Accuracy_score: ", accuracy_score(y_pred_test,y_test))
```

TASK-2 correct workflow

```python
x_train1,x_test1,y_train1,y_test1= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train_scaled= scale.fit_transform(x_train1)
x_test_scaled= scale.transform(x_test1)
```

```python
lr_2= LogisticRegression()
```

```python
lr_2.fit(x_train_scaled,y_train1)
```

```python
y_pred_train1= lr_2.predict(x_train_scaled)
y_pred_test1= lr_2.predict(x_test_scaled)
```

```python
print("Training_Accuracy_score: ", accuracy_score(y_pred_train1,y_train1))
print("Test_Accuracy_score: ", accuracy_score(y_pred_test1,y_test1))
```

TASK-3 concept understanding

1. What is Data Leakage?

Data leakage occurs when information from outside the training dataset (especially from the test set or future data) is used to build the model.

👉 This gives the model unfair knowledge, leading to overly high accuracy during training/testing, but poor real-world performance.

Example:
Using future sales data to predict past sales.

2. Why should preprocessing be done only on training data?

Preprocessing steps (like scaling, encoding, imputation) must be fit only on training data because:

If you use the full dataset (train + test), information from the test set leaks into the model.
This makes evaluation unrealistic since the model has already “seen” patterns from test data.

👉 Correct approach:

Fit preprocessing on training data
Apply (transform) the same steps to test data

3. Can a feature derived from the target variable be used as input?

❌ No, it should not be used.

Reason:

It introduces data leakage because the model indirectly gets access to the answer (target).
This leads to artificially high accuracy but fails in real-world predictions.

Example:
If predicting whether a student passes, using a feature like “final result score” (derived from pass/fail) is invalid.
