Rule-Based AI

Rule-based AI systems operate using predefined rules written by humans. These rules follow an “if–then” logic:

Example: IF temperature > 30°C THEN turn on fan

Machine learning systems learn patterns from data instead of explicit rules. They improve performance over time by analyzing examples.

Spam Email Filters (Basic Version)
Early email systems used rules like:
IF subject contains “win money” → mark as spam
Works well for simple patterns but fails for new tricks.

Fraud Detection in Banking
Fraud patterns constantly change and are complex.
Rule-based system would fail because:
You cannot manually define all possible fraud scenarios.

Machine Learning:

Learns from transaction data
Detects unusual behavior (e.g., sudden large transactions in a new location)
Continuously improves over time
